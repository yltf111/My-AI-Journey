```python
#循环操作---for
for i in range(5):
    print(i)
```

```python
#循环操作---while
sum = 0
n = 99
while n > 0:
    sum = sum + n
    n = n - 1
print(sum)
```

```python
#break、continue、pass的作用
#break语句可以跳出 for 和 while 的循环体
n = 1
while n <= 100:
    if n > 10:
        break
    print(n)
    n += 1
```

```python
#continue语句跳过当前循环，直接进行下一轮循环
n = 1
while n < 10:
    n = n + 1
    if n % 2 == 0:
        continue
    print(n)
```

```python
#pass是空语句，一般用做占位语句，不做任何事情
for letter in 'Room':
    if letter == 'o':
        pass
        print('pass')
    print(letter)
```

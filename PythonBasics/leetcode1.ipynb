```python
#给你一个整数 n 。按下述规则生成一个长度为 n + 1 的数组 nums ：

#nums[0] = 0
#nums[1] = 1
#当 2 <= 2 * i <= n 时，nums[2 * i] = nums[i]
#当 2 <= 2 * i + 1 <= n 时，nums[2 * i + 1] = nums[i] + nums[i + 1]
#返回生成数组 nums 中的 最大 值。
class Solution:
    def getMaximumGenerated(self, n):
        if n == 0:
            return 0
        if n == 1:
            return 1
        
        nums = [0] * (n + 1)
        nums[1] = 1
        
        for i in range(1, n // 2 + 1):
            j1 = 2 * i
            if j1 <= n:
                nums[j1] = nums[i]
            j2 = 2 * i + 1
            if j2 <= n:
                nums[j2] = nums[i] + nums[i + 1]
        
        return max(nums)
```

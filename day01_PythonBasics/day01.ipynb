```python
# 创建一个集合set_v
set_v = set([1,2,3])
print(2 in set_v)    #2在集合set_v当中
print(2 not in set_v)    #2不在集合set_v当中   
```

```python
# 创建一个元组tuple_v
tuple_v = (1,2,3)    #也可以是= 1,2,3
print(2 in tuple_v)
print(2 not in tuple_v)
```
